```python
import os
import cv2
import numpy as np

def extract_frames_from_video(video_path, output_folder):
    # Open video using OpenCV
    video_capture = cv2.VideoCapture(video_path)
    video_name = os.path.basename(video_path).split('.')[0]
    frame_count = 0
    frame_number = 0

    # Create the output folder for the video
    video_output_folder = os.path.join(output_folder, video_name)
    os.makedirs(video_output_folder, exist_ok=True)

    while True:
        ret, frame = video_capture.read()

        # If the frame is read correctly, ret is True
        if not ret:
            break

        frame_count += 1

        # Save every 10th frame
        if frame_count % 10 == 0:
            # Resize frame to 224x224
            resized_frame = cv2.resize(frame, (224, 224))

            # Save the frame with the format "framenumber.png"
            frame_filename = os.path.join(video_output_folder, f"frame{frame_number}.png")
            cv2.imwrite(frame_filename, resized_frame)
            frame_number += 1

    video_capture.release()

def process_category_videos(input_folder, output_folder):
    # Get list of subdirectories (categories)
    categories = os.listdir(input_folder)

    for category in categories:
        category_path = os.path.join(input_folder, category)

        if os.path.isdir(category_path):
            # Create a folder for this category in the output directory
            category_output_folder = os.path.join(output_folder, category)
            os.makedirs(category_output_folder, exist_ok=True)

            # Process each video in the category
            videos = os.listdir(category_path)
            for video in videos:
                video_path = os.path.join(category_path, video)

                if os.path.isfile(video_path) and video.endswith('.mp4'):
                    extract_frames_from_video(video_path, category_output_folder)

if __name__ == "__main__":
    input_dir = r"/home/fast/Downloads/Anomaly-Detection-Dataset/test"
    output_dir = r"/home/fast/Downloads/Anomaly-Detection-Dataset/testframes"

    # Process all categories in the input directory
    process_category_videos(input_dir, output_dir)
    print("Frame extraction complete.")
```

```python
import os
import random
import cv2
import numpy as np
from tqdm import tqdm

# Define the augmentation methods
def flip_frames(frames):
    return [cv2.flip(frame, 1) for frame in frames]

def add_noise(frames):
    return [
        cv2.add(frame, np.random.randint(0, 50, frame.shape, dtype='uint8'))
        for frame in frames
    ]

def change_brightness(frames):
    alpha = random.uniform(0.7, 1.3)  # Random brightness multiplier
    return [cv2.convertScaleAbs(frame, alpha=alpha, beta=0) for frame in frames]

def rotate_frames(frames):
    angle = random.choice([90, 180, 270])
    return [cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE) if angle == 90 else
            cv2.rotate(frame, cv2.ROTATE_180) if angle == 180 else
            cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE) for frame in frames]

# List of augmentation methods
AUGMENTATIONS = [flip_frames, add_noise, change_brightness, rotate_frames]

# Augment videos in the given directory
def augment_videos(base_dir, target_count=450):
    categories = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

    for category in categories:
        category_path = os.path.join(base_dir, category)
        video_folders = [d for d in os.listdir(category_path) if os.path.isdir(os.path.join(category_path, d))]

        print(f"Processing category: {category} (Videos: {len(video_folders)})")

        if len(video_folders) >= target_count:
            print(f"Skipping category {category}, already has {len(video_folders)} videos.")
            continue

        while len(video_folders) < target_count:
            # Select a random video folder
            source_folder = random.choice(video_folders)
            source_path = os.path.join(category_path, source_folder)

            # Load frames from the source folder
            frames = []
            for frame_file in sorted(os.listdir(source_path)):
                frame_path = os.path.join(source_path, frame_file)
                frame = cv2.imread(frame_path)
                if frame is not None:
                    frames.append(frame)

            if not frames:
                print(f"No frames found in {source_folder}, skipping.")
                continue

            # Apply a random augmentation
            augmentation = random.choice(AUGMENTATIONS)
            augmented_frames = augmentation(frames)

            # Create a new folder for the augmented video
            new_folder_name = f"{source_folder}_aug{random.randint(1000, 9999)}"
            new_folder_path = os.path.join(category_path, new_folder_name)
            os.makedirs(new_folder_path, exist_ok=True)

            # Save the augmented frames
            for i, frame in enumerate(augmented_frames):
                frame_name = f"frame{i:03d}.png"
                cv2.imwrite(os.path.join(new_folder_path, frame_name), frame)

            video_folders.append(new_folder_name)  # Update the list of videos
            print(f"Augmented video saved to: {new_folder_name}")

        print(f"Category {category} now has {len(video_folders)} videos.")

# Main function
if __name__ == "__main__":
    base_dir = r"E:\Output_Dir"
    augment_videos(base_dir)
```

```python
import os
import random
import cv2
import numpy as np
from tqdm import tqdm

# Define the augmentation methods
def flip_frames(frames):
    return [cv2.flip(frame, 1) for frame in frames]

def add_noise(frames):
    return [
        cv2.add(frame, np.random.randint(0, 50, frame.shape, dtype='uint8'))
        for frame in frames
    ]

def change_brightness(frames):
    alpha = random.uniform(0.7, 1.3)  # Random brightness multiplier
    return [cv2.convertScaleAbs(frame, alpha=alpha, beta=0) for frame in frames]

def rotate_frames(frames):
    angle = random.choice([90, 180, 270])
    return [cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE) if angle == 90 else
            cv2.rotate(frame, cv2.ROTATE_180) if angle == 180 else
            cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE) for frame in frames]

# List of augmentation methods
AUGMENTATIONS = [flip_frames, add_noise, change_brightness, rotate_frames]

# Augment videos in the given directory
def augment_videos(base_dir, target_count=120):
    categories = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

    for category in categories:
        category_path = os.path.join(base_dir, category)
        video_folders = [d for d in os.listdir(category_path) if os.path.isdir(os.path.join(category_path, d))]

        print(f"Processing category: {category} (Videos: {len(video_folders)})")

        if len(video_folders) >= target_count:
            print(f"Skipping category {category}, already has {len(video_folders)} videos.")
            continue

        iteration = 1
        while len(video_folders) < target_count:
            # Select a random video folder
            source_folder = random.choice(video_folders)
            source_path = os.path.join(category_path, source_folder)

            # Load frames from the source folder
            frames = []
            for frame_file in sorted(os.listdir(source_path)):
                frame_path = os.path.join(source_path, frame_file)
                frame = cv2.imread(frame_path)
                if frame is not None:
                    frames.append(frame)

            if not frames:
                print(f"No frames found in {source_folder}, skipping.")
                continue

            # Apply a random augmentation
            augmentation = random.choice(AUGMENTATIONS)
            print(f"Augmenting video: {source_folder} using {augmentation.__name__}")
            augmented_frames = augmentation(frames)

            # Create a new folder for the augmented video
            new_folder_name = f"{source_folder}_aug{iteration}"
            iteration += 1
            new_folder_path = os.path.join(category_path, new_folder_name)
            os.makedirs(new_folder_path, exist_ok=True)

            # Save the augmented frames
            for i, frame in enumerate(augmented_frames):
                frame_name = f"frame{i:03d}.png"
                cv2.imwrite(os.path.join(new_folder_path, frame_name), frame)

            video_folders.append(new_folder_name)  # Update the list of videos
            print(f"Augmented video saved to: {new_folder_name}")

        print(f"Category {category} now has {len(video_folders)} videos.")

# Main function
if __name__ == "__main__":
    base_dir = r"/home/fast/Downloads/Anomaly-Detection-Dataset/trainframes"
    augment_videos(base_dir)
```

```python
import os
import random
import shutil

# Function to split the dataset into train and test
def split_dataset(base_dir, test_dir, test_ratio=0.2):
    if not os.path.exists(test_dir):
        os.makedirs(test_dir)

    categories = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

    for category in categories:
        category_path = os.path.join(base_dir, category)
        test_category_path = os.path.join(test_dir, category)

        if not os.path.exists(test_category_path):
            os.makedirs(test_category_path)

        video_folders = [d for d in os.listdir(category_path) if os.path.isdir(os.path.join(category_path, d))]
        test_count = int(len(video_folders) * test_ratio)

        print(f"Processing category: {category} (Total videos: {len(video_folders)}, Test videos: {test_count})")

        test_videos = random.sample(video_folders, test_count)

        for video in test_videos:
            source_path = os.path.join(category_path, video)
            destination_path = os.path.join(test_category_path, video)

            shutil.move(source_path, destination_path)
            print(f"Moved video {video} to test folder.")

    print("Dataset splitting complete.")

# Main function
if __name__ == "__main__":
    base_dir = r"E:\Output_Dir"
    test_dir = r"E:\Output_Dir-Test"

    split_dataset(base_dir, test_dir)
```
